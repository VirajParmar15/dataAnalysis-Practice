Measure of central tendency
1. mean
2. median
3. mode

```python
ages=[23,24,32,45,12,43,67,45,32,56,32,102]
```

```python
import numpy as np
import statistics
print(np.mean(ages))
print(np.median(ages))
print(statistics.mean(ages))
print(statistics.median(ages))
print(statistics.mode(ages))
```

```python
import seaborn as sns
sns.boxplot(ages)
```

5 number summary

```python
import numpy as np
q1,q3=np.percentile(ages,[25,75])
print(q1)
print(q3)
```

```python
## to check outliers[Lower Fence - Upper Fence]
iqr=q3-q1
lower_fence=q1-1.5*(iqr)
upper_fence=q3+1.5*(iqr)
print(lower_fence,upper_fence)
```

Measure of dispersion
1. standard deviation
2. variance

```python
statistics.variance(ages)
```

```python
np.std(ages)
```

```python
def variance(data):
  n=len(ages)
  mean=sum(data)/n
  deviation=[(x - mean) ** 2 for x in data]
  variance=sum(deviation)/n
  return variance
```

```python
variance(ages)
```

```python
import math
math.sqrt(statistics.pvariance(ages))
```

check whether distribution is normal distribution or not

```python
###if you want to check whether feature is gaussian or normal distributed
import matplotlib.pyplot as plt
import scipy.stats as stat
import pylab
def plot_data(sample):
  plt.figure(figsize=(10,6))
  plt.subplot(1,2,1)
  sns.histplot(sample)
  plt.subplot(1,2,2)
  stat.probplot(sample,dist='norm',plot=pylab)
  plt.show()
```

```python
##create  a normal distribution data
s=np.random.normal(0.5,0.2,1000)
plot_data(s)
```

```python
plot_data(df['sepal_width'])
```

```python
mu, sigma = 3., 1.
sample = np.random.lognormal(mu,sigma,100)
```

```python
def plot_data_log(sample):
  plt.figure(figsize=(10,6))
  plt.subplot(1,2,1)
  sns.histplot(sample)
  plt.subplot(1,2,2)
  stat.probplot(np.log(sample),dist='norm',plot=pylab)
  plt.show()
```

```python
plot_data_log(sample)
```
