```python
import pandas as pd
df = pd.read_csv("/content/drive/MyDrive/DATASETS/Top Indian Places to Visit.csv")
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
 df['Significance'].value_counts()
```

```python
selected_types = ['Historical','Religious','Nature','Recreational','Wildlife']
filtered_df = df[df['Significance'].isin(selected_types)]
types = filtered_df['Significance'].value_counts()
plt.pie(types,labels=types.index,autopct='%1.1f%%',startangle=90)
plt.axis('equal')
plt.show()
```

```python
types = df['Airport with 50km Radius'].value_counts()
types
```

```python
value_counts = df['Airport with 50km Radius'].value_counts()
plt.figure(figsize=(7,4.5))
plt.bar(value_counts.index, value_counts.values, color='skyblue')

plt.xlabel('Airport with 50km Radius')
plt.ylabel('Counts')
plt.title('Counts of Airport with 50km Radius')

plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()
```

```python
data_counts = df['DSLR Allowed'].value_counts()
plt.figure(figsize=(7,4.5))
plt.bar(data_counts.index,data_counts.values, color='lightgreen')
plt.xlabel('Dslr allowed')
plt.ylabel('counts')
plt.title('counts of dslr allowed')

plt.tight_layout()
plt.show()
```
