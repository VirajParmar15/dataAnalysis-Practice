```python
df_country=pd.read_excel('/content/drive/MyDrive/Country-Code.xlsx')
df_country
```

```python
df.columns
```

```python
final_df=pd.merge(df,df_country,on='Country Code',how='left')
```

```python
final_df.head(2)
```

```python
final_df.dtypes
```

```python
final_df.columns
```

```python
country_names=final_df.Country.value_counts().index
```

```python
country_val=final_df.Country.value_counts().values
```

```python
##pie chart-top 3
plt.pie(country_val[:3],labels=country_names[:3],autopct='%1.2f%%')
```

Observation:Zomato maximum reocrds or transaction are from India
After that usa and then united kingdom

```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating count'})
```

```python
ratings
```

##observation
* 1.when rating are between 4.5 to 4.9 ---> excellent
* 2.when rating are between 4.0 to 3.4 ---> very good
* 3.when rating are between 3.5 to 3.9 ---> good
* 4.when rating are between 3.0 to 3.4 ---> average
* 5.when rating are between 2.5 to 2.9 ---> average
* 6.when rating are between 2.0 to 2.4 ---> poor

```python
ratings.head()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(x="Aggregate rating", y="Rating count",hue='Rating color',data=ratings,palette=['white','red','orange','green','green'])
```

##Observation
* 1.Note rated count is very high
* 2.Maximum number of rating are between 2.5 to 3.4

```python
## Count plot
sns.countplot(x='Rating color',data=ratings,palette=['blue','red','orange','yellow','green','green'])
```
